```python
import numpy as np
import scipy.signal as ss
import matplotlib.pyplot as plt
```

```python
def minmax(x):
    """min max normalizes the given array"""
    return (x - np.min(x))/(np.max(x)-np.min(x))

B_old = [0.049922035, -0.095993537, 0.050612699, -0.004408786]
A_old = [1, -2.494956002,   2.017265875,  -0.522189400]

B_new = [0.00512885, 0.08889104, 0.08889104, 0.00512885]
A_new = [1, -2.494956002,   2.017265875,  -0.522189400]

t = 10000
wn = np.random.normal(loc=1,scale=0.5,size=t+2000)
invfn_old = minmax(ss.lfilter(B_old, A_old, wn)[2000:])+0.5
invfn_new_filt = minmax(ss.lfilter(B_new, A_new, wn)[2000:])+0.5
invfn_new = (invfn_new_filt**5)-np.mean((invfn_new_filt**5))+1
invfn_new[invfn_new<0] = 0
```

```python
plt.plot(np.linspace(0,10,t),invfn_new)
plt.xlabel('Time (s)')
plt.ylabel('Normalized Signal')
plt.title('Firing Rate')
```
